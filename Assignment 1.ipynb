```python
import pandas as pd
import numpy as np

pd.DataFrame({
    "Product":["Laptop","Mobile","Tablet"],
    "Price":[50000,20000,15000]
}).to_csv("products.csv",index=False)

df = pd.read_csv("products.csv")
avg = np.mean(df["Price"])

print("Average Price =", avg)
print(df[df["Price"] > avg])
```

```python
import pandas as pd
import numpy as np

pd.DataFrame({
    "Day":["Mon","Tue","Wed"],
    "Temperature":[18,25,32]
}).to_csv("temperature.csv",index=False)

df = pd.read_csv("temperature.csv")

print("Highest =", np.max(df["Temperature"]))
print("Lowest =", np.min(df["Temperature"]))
print("Average =", np.mean(df["Temperature"]))

def temp(t):
    if t < 20:
        return "Cold"
    elif t <= 30:
        return "Moderate"
    else:
        return "Hot"

df["Category"] = df["Temperature"].apply(temp)
print(df)
```
